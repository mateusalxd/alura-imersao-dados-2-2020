## Desafio - 2
Criar uma função para mostrar o gráfico de maneira mais fácil

```
def exibir_grafico(data=None, x=None, y=None, pkind=None, title=None, figsize=None, palette=None, hue=None, auto_show=True, **kwargs):
  # define o tema
  sns.set_theme()

  # verifica se o pkind foi informado
  if pkind != None:

    # verifica se o figsize foi informado
    if figsize != None:
      # define o tamanho da figura do gráfico
      plt.figure(figsize=figsize)
    

    # define as variáveis de labels
    xlabel = None
    ylabel = None

    # tenta recuperar o parâmetro de label do eixo X
    try:
      xlabel = kwargs.pop("xlabel")
    except:
      # verifica se x é um texto
      if type(x) == str:
        xlabel = x
    
    # tenta recuperar o parâmetro de label do eixo Y
    try:
      ylabel = kwargs.pop("ylabel")
    except:
      # verifica se y é um texto
      if type(y) == str:
        ylabel = y
    

    # variável que irá armazenar os eixos
    eixos = None

    # se o pkind for bar
    if pkind == "barplot":
      # gera o gráfico com os parâmetros
      eixos = sns.barplot(x=x, y=y, data=data, palette=palette, **kwargs)
    # senão se o pkind for boxplot
    elif pkind == "boxplot":
      # gera o gráfico com os parâmetros
      eixos = sns.boxplot(x=x, y=y, data=data, palette=palette, hue=hue, **kwargs)
    # senão se o pkind for histplot      
    elif pkind == "histplot":
      # de para
      de_para_label_y = {"Count":"Contagem", 
                         "Frequency": "Frequência", 
                         "Density": "Densidade", 
                         "Probability": "Probabilidade"}
      # gera o gráfico com os parâmetros
      eixos = sns.histplot(x=x, data=data, palette=palette, hue=hue, **kwargs)
      # define o label para o eixo Y
      ylabel = de_para_label_y[eixos.get_ylabel()]
    # senão se o pkind for scatterplot
    elif pkind == "scatterplot":
      # gera o gráfico com os parâmetros
      eixos = sns.scatterplot(x=x, y=y, data=data, palette=palette, hue=hue, **kwargs)   
    # senão se o pkind for paitplot
    elif pkind == "pairplot":
      # gera o gráfico com os parâmetros
      eixos = sns.pairplot(data=data, palette=palette, hue=hue, **kwargs)
    # senão se o pkind for heatmap
    elif pkind == "heatmap":
      # gera o gráfico com os parâmetros
      eixos = sns.heatmap(data=data, **kwargs)         
    

    # verifica se foram retornados os eixos
    if eixos != None and pkind != "pairplot":
      # verifica se existe o label de x
      if xlabel != None:
        # define o label de x
        eixos.set(xlabel=xlabel)
      
      # verifica se existe o label de y
      if ylabel != None:
        # define o label de y
        eixos.set(ylabel=ylabel)


    # verifica se o title foi informado
    if title != None:
      # define o título do gráfico
      plt.title(title)
    

    # verifica se é para exibir automaticamente
    if auto_show:
      # exibe o gráfico
      plt.show()
      # não retorna nada
      return None
    else:
      # retorna os eixos
      return eixos

  # senão
  else:
    # lança uma exceção
    raise Exception("Não foi informado o tipo de gráfico")
```

```
import numpy as np
from statistics import mean 
```

```
x_treino_manual = pd.concat([x_treino, y_treino],axis=1)
```

```
def retornar_valor(grupo, intervalo, valor):
  resultado = grupo[intervalo.contains(valor)].mean()
  return mean(resultado.values) if not np.isnan(mean(resultado.values)) else valor
```

```
def prever_valor(x):
  valores = []
  for linha in x.fillna(0.0).to_numpy():
    parciais = []
    parciais.append(retornar_valor(grupo_cn, intervalo_grupo_cn, linha[0]))
    parciais.append(retornar_valor(grupo_ch, intervalo_grupo_ch, linha[1]))
    parciais.append(retornar_valor(grupo_lc, intervalo_grupo_lc, linha[2]))
    parciais.append(retornar_valor(grupo_rd, intervalo_grupo_rd, linha[3]))
    valores.append(mean(parciais))
  return valores
```

```
intervalos_cn = np.arange(x_treino_manual["NU_NOTA_CN"].min(), x_treino_manual["NU_NOTA_CN"].max(), 25.0)
bin_cn = pd.cut(x_treino_manual["NU_NOTA_CN"], intervalos_cn)
grupo_cn = x_treino_manual[["NU_NOTA_CN", "NU_NOTA_MT"]].groupby(bin_cn)["NU_NOTA_MT"].median().reset_index()
intervalo_grupo_cn = pd.arrays.IntervalArray(grupo_cn['NU_NOTA_CN'])
```

```
intervalos_ch = np.arange(x_treino_manual["NU_NOTA_CH"].min(), x_treino_manual["NU_NOTA_CH"].max(), 25.0)
bin_ch = pd.cut(x_treino_manual["NU_NOTA_CH"], intervalos_ch)
grupo_ch = x_treino_manual[["NU_NOTA_CH", "NU_NOTA_MT"]].groupby(bin_ch)["NU_NOTA_MT"].median().reset_index()
intervalo_grupo_ch = pd.arrays.IntervalArray(grupo_ch['NU_NOTA_CH'])
```

```
intervalos_lc = np.arange(x_treino_manual["NU_NOTA_LC"].min(), x_treino_manual["NU_NOTA_LC"].max(), 25.0)
bin_lc = pd.cut(x_treino_manual["NU_NOTA_LC"], intervalos_lc)
grupo_lc = x_treino_manual[["NU_NOTA_LC", "NU_NOTA_MT"]].groupby(bin_lc)["NU_NOTA_MT"].median().reset_index()
intervalo_grupo_lc = pd.arrays.IntervalArray(grupo_lc["NU_NOTA_LC"])
```

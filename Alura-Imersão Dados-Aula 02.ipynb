### Desafio - 2
Criar uma função para mostrar o gráfico de maneira mais fácil

```
def exibir_grafico(data, x, y, kind=None, title=None, figsize=None, palette=None, hue=None, **kwargs):
  # define o tema
  sns.set_theme()
  # verifica se o kind foi informado
  if kind != None:
    # verifica se o figsize foi informado
    if figsize != None:
      # define o tamanho da figura do gráfico
      plt.figure(figsize=figsize)
    

    # se o kind for bar
    if kind == "bar":
      # gera o gráfico com os parâmetros
      sns.barplot(x=x, y=y, data=data, palette=palette, **kwargs)
    # senão se o kind for boxplot
    elif kind == "boxplot":
      # gera o gráfico com os parâmetros
      sns.boxplot(x=x, y=y, data=data, palette=palette, hue=hue, **kwargs)

    # verifica se o title foi informado
    if title != None:
      # define o título do gráfico
      plt.title(title)
    
    # exibe o gráfico
    plt.show()
  # senão
  else:
    # lança uma exceção
    raise Exception("Não foi informado o tipo de gráfico")
```

### Desafio - 1
Comparar os alunos menores de idade, separando por Estado

```
# conta a quantidade de menores de idade por estado, ordenando pelo estado,
# depois transforma a Series em um DataFrame, passando o índice para a coluna
# e finaliza renomeando todas as colunas
qtde_menor_idade_por_estado = dados.query("NU_IDADE < 18")["SG_UF_RESIDENCIA"] \
  .value_counts() \
  .sort_index() \
  .reset_index() \
  .rename(columns={"index":"SG_UF_RESIDENCIA", 
                   "SG_UF_RESIDENCIA":"NU_QTDE_MENOR_IDADE"})

# exibe o gráfico usando a função criada
exibir_grafico(data=qtde_menor_idade_por_estado,
               x="SG_UF_RESIDENCIA",
               y="NU_QTDE_MENOR_IDADE",
               kind="bar",
               title="Quantidade de menores de idade por Estado",
               figsize=(18,10),
               palette="Set2")
```

```
# lista com as colunas relacionadas a presença
presencas = ["TP_PRESENCA_CN", "TP_PRESENCA_CH", "TP_PRESENCA_LC", "TP_PRESENCA_MT"]
# conta as notas iguais a zero, considerando os tipos de presenças, normalizando
# a contagem, em seguida renomeia a coluna da contagem e depois transforma a 
# Series em um DataFrame
presencas_com_nota_zerada = dados.query("NU_NOTA_TOTAL == 0")[presencas] \
  .value_counts(normalize=True) \
  .rename("NU_QTDE") \
  .reset_index()

# exibe o DataFrame, ajustando os valores para ficar mais fácil de entender
presencas_com_nota_zerada.replace({0: "Faltou à prova", 1: "Presente na prova", 2:"Eliminado na prova"})
```

```
# lê o arquivo em partes
# https://pandas.pydata.org/pandas-docs/stable/user_guide/io.html#io-chunking
fonte_completa = "DADOS/MICRODADOS_ENEM_2019.csv"
leitor = pd.read_csv(fonte_completa, 
                     sep=",", 
                     encoding="ISO-8859-1", 
                     chunksize=1000000, 
                     delimiter=";")
# array para armazenar as partes
partes = [ ]
# total de linhas
contador = 0

# percorre todas as partes do arquivo
for parte in leitor:
  # incrementa o contador com a quantidade de linhas
  contador += parte.shape[0]
  # adiciona os dados de acordo com o filtro
  partes.append(parte.query("NU_IDADE == 13 or NU_IDADE == 14"))

# junta as partes em um único DataFrame
dados_13_14_anos = pd.concat(partes)
# imprime o total de linhas
print(f"Total de linhas do arquivo completo: {contador}\tLinhas selecionadas: {dados_13_14_anos.shape[0]}")
```

```
# gera gráfico boxplot das notas dos alunos de 13 e 14 anos
dados_13_14_anos[provas].plot.box(grid=True, figsize=(12,8))
# define o título do gráfico
plt.title("Distribuição das Notas das Provas em Alunos de 13 e 14 anos")
# exibe o gráfico
plt.show()
```

```
# gera o gráfico de barra da quantidade de alunos por estado
dados_13_14_anos["SG_UF_RESIDENCIA"].value_counts().plot.bar(grid=True, figsize=(12,8))
# define o título do gráfico
plt.title("Quantidade de Alunos de 13 e 14 anos por Estado")
# exibe o gráfico
plt.show()
```

```
# gera o gráfico de pizza por alunos treineiros ou não
dados_13_14_anos["IN_TREINEIRO"].value_counts().plot.pie(figsize=(8,8))
# define o título do gráfico
plt.title("Divisão dos Aluno de 13 e 14 anos por treineiros")
# exibe o gráfico
plt.show()
```

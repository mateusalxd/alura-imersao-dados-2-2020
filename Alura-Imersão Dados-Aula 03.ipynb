## Desafio - 2
Criar uma função para mostrar o gráfico de maneira mais fácil

```
def exibir_grafico(data, x=None, y=None, kind=None, title=None, figsize=None, palette=None, hue=None, auto_show=True, **kwargs):
  # define o tema
  sns.set_theme()
  # verifica se o kind foi informado
  if kind != None:
    # verifica se o figsize foi informado
    if figsize != None:
      # define o tamanho da figura do gráfico
      plt.figure(figsize=figsize)
    

    # define as variáveis de labels
    xlabel = None
    ylabel = None

    # tenta recuperar o parâmetro de label do eixo X
    try:
      xlabel = kwargs.pop("xlabel")
    except:
      xlabel = x
    
    # tenta recuperar o parâmetro de label do eixo Y
    try:
      ylabel = kwargs.pop("ylabel")
    except:
      ylabel = y
    

    # variável que irá armazenar os eixos
    eixos = None

    # se o kind for bar
    if kind == "barplot":
      # gera o gráfico com os parâmetros
      eixos = sns.barplot(x=x, y=y, data=data, palette=palette, **kwargs)
    # senão se o kind for boxplot
    elif kind == "boxplot":
      # gera o gráfico com os parâmetros
      eixos = sns.boxplot(x=x, y=y, data=data, palette=palette, hue=hue, **kwargs)
    # senão se o kind for histplot      
    elif kind == "histplot":
      # de para
      de_para_label_y = {"Count":"Contagem", 
                         "Frequency": "Frequência", 
                         "Density": "Densidade", 
                         "Probability": "Probabilidade"}
      # gera o gráfico com os parâmetros
      eixos = sns.histplot(x=x, data=data, palette=palette, hue=hue, **kwargs)
      # define o label para o eixo Y
      ylabel = de_para_label_y[eixos.get_ylabel()]
    # senão se o kind for scatterplot
    elif kind == "scatterplot":
      # gera o gráfico com os parâmetros
      eixos = sns.scatterplot(x=x, y=y, data=data, palette=palette, hue=hue, **kwargs)   
    # senão se o kind for paitplot
    elif kind == "pairplot":
      # gera o gráfico com os parâmetros
      eixos = sns.pairplot(data=data, palette=palette, hue=hue, **kwargs)
    # senão se o kind for heatmap
    elif kind == "heatmap":
      # gera o gráfico com os parâmetros
      eixos = sns.heatmap(data=data, **kwargs)         
    

    # verifica se foram retornados os eixos
    if eixos != None and kind != "pairplot":
      # define os labels dos eixos, caso tenham sido passados como parâmetros
      eixos.set(xlabel=xlabel, ylabel=ylabel)


    # verifica se o title foi informado
    if title != None:
      # define o título do gráfico
      plt.title(title)
    

    # verifica se é para exibir automaticamente
    if auto_show:
      # exibe o gráfico
      plt.show()
      # não retorna nada
      return None
    else:
      # retorna os eixos
      return eixos

  # senão
  else:
    # lança uma exceção
    raise Exception("Não foi informado o tipo de gráfico")
```

### Desafio 1
Plotar a média, mediana e moda nas notas de LC e MT

```
def grafico_desafio_1(data, x, xlabel, title):
  # cria a base do gráfico
  exibir_grafico(data=data, 
                x=x, 
                kind="histplot", 
                title=title,
                figsize=(12,8),
                xlabel=xlabel,
                auto_show=False)

  # calcula a média, mediana e moda
  media = data[x].mean()
  mediana = data[x].median()
  moda = data[x].mode()

  # imprime os valores
  print(f"Média (vermelho): {media:.1f}, Mediana (amarelo): {mediana:.1f}, Moda (verde): {moda.values}")

  # adiciona linhas verticais correspondente a média, mediana e moda
  plt.axvline(media, color="red")
  plt.axvline(mediana, color="yellow")
  for valor in moda.values:
    plt.axvline(valor, color="green")

  # exibe o gráfico
  plt.show()
```

```
grafico_desafio_1(dados_sem_notas_zero, "NU_NOTA_TOTAL", "Nota Total", "Histograma da Nota Total")
grafico_desafio_1(dados_sem_notas_zero, "NU_NOTA_MT", "Nota de Matemática", "Histograma da Nota de Matemática")
grafico_desafio_1(dados_sem_notas_zero, "NU_NOTA_LC", "Nota de Linguagem e Código", "Histograma da Nota de Linguagem e Código")
```
